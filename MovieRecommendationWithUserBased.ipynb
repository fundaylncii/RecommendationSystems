```python
## Kullanıcı benzerlikleri üzerinden kullanıcı film önerileri oluşturma
## DataSet Url: https://grouplens.org/datasets/movielens/
```

```python
def create_user_movie_df():
  import pandas as pd
  movie = pd.read_csv("/content/movie.csv")
  rating = pd.read_csv("/content/rating.csv")

  ## read edilen iki tablo movieId'e göre birleştirilir
  df = movie.merge(rating, how="left", on="movieId")

  ## film bazında yorum sayılarının yer aldığı df oluşturulur
  comment_counts = pd.DataFrame(df["title"].value_counts())

  ## 1000 den az yorum alan filmler tespit edilir
  rare_movies = comment_counts[comment_counts["count"] <= 1000].index

  ## 1000 den az yorum alan filmlerin olmadığı bir df oluşturulur
  common_movies = df[~df["title"].isin(rare_movies)]

  ## satırda userId kolonlarda film isimleri ve değerlerde de puanlar olacak şekilde pivot table oluşturulur
  user_movies_df = common_movies.pivot_table(index=["userId"], columns=["title"], values="rating")
  return user_movies_df
```

```python
user_movies_df = create_user_movie_df()
user_movies_df.head(10)
```

```python
## rastgele kullanıcı seçimi
import pandas as pd
random_user = int(pd.Series(user_movies_df.index).sample(1).values[0])
print(random_user)
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.width",500)
pd.set_option("display.expand_frame_repr",False)

## öneri yapılacak kullanıcın izlediği filmlerin belirlenmesi
random_user_df = user_movies_df[user_movies_df.index == random_user]
print(random_user_df)
```

```python
## izlenen filmlerin getirilmesi

movies_watched = random_user_df.columns[random_user_df.notna().any()].tolist()
print(movies_watched)
```

```python
## gelen datanı ispatı

user_movies_df.loc[user_movies_df.index == random_user, user_movies_df.columns == "Birdcage, The (1996)"]
```

```python
## rondom userın izlediği filmler ile df oluşturulur

movies_watched_df = user_movies_df[movies_watched]
movies_watched_df.head(10)
```

```python
## her userın kaç tane film izlediğinin tespiti
user_movies_count = movies_watched_df.T.notnull().sum()
print(user_movies_count)
```

```python
user_movies_count = user_movies_count.reset_index()
print(user_movies_count)
```

```python
user_movies_count.columns = ["userId", "movie_count"]
print(user_movies_count)
```

```python
## random user ile %60 oranında aynı film izleyenler
perc = len(movies_watched) * 60 / 100
user_movies_count[user_movies_count["movie_count"] > perc].sort_values("movie_count", ascending=False)
```

```python
## random user ile tüm filmleri aynı izleyen kullanıcıların sayısı
user_movies_count[user_movies_count["movie_count"] == 14].count()
```

```python
## benzerlik taşıyan userlerin id si
users_same_movie = user_movies_count[user_movies_count["movie_count"] > perc]["userId"]
print(users_same_movie)
```

```python
## oluştulan df lerin birleştirilmesi

final_df = pd.concat([movies_watched_df[movies_watched_df.index.isin(users_same_movie)],
                      random_user_df[movies_watched]])

final_df.head(10)
```

```python
## belirlenen userların davranış benzerlikleri tespit edilir.
## kullanıcıları kolonlara alıp corelasyonu alınıp pivot yapıldıktan sonra sıralanıp tekrar eden değerler çıkartılır

corr_df = final_df.T.corr().unstack().sort_values().drop_duplicates()
corr_df.head(10)
```

```python
corr_df = pd.DataFrame(corr_df, columns=["corr"])
corr_df.index.names=["user_id_1","user_id_2"]
corr_df = corr_df.reset_index()
corr_df.head(10)
```

```python
## random user ile %65 pozitif korelasyon gösteren kullanıcıların tespiti ( yani %65 oranında aynı davranışı sergilerler)
top_user = corr_df[(corr_df["user_id_1"] == random_user) & (corr_df["corr"] >= 0.65)][["user_id_2", "corr"]].reset_index(drop=True)
top_user.head(10)
```

```python
## kullanıcıların sıralanması
top_user = top_user.sort_values("corr", ascending=False)
top_user.rename(columns={"user_id_2": "userId"}, inplace=True)
print(top_user)
```

```python
## rating df ile top user df in birleştirilmesi
top_user_rating = top_user.merge(rating[["userId","movieId","rating"]], how="inner")


## random userın df içerisinden çıkartılması
top_user_rating = top_user_rating[top_user_rating["userId"] != random_user]
print(top_user_rating)
```

```python
## Weighted Average Recommendation Score Hesaplaması:
## corelasyon ve rate etkisini aynı şekilde hesaplayabilmek için yeni bir değişken oluşturulur.

top_user_rating["weighted_rating"] = top_user_rating["corr"] * top_user_rating["rating"]
print(top_user_rating)

## rating değerleri corelasyonlara göre düzenlendi
```

```python
recommendation_df =top_user_rating.groupby("movieId").agg({"weighted_rating": "mean"})
recommendation_df = recommendation_df.reset_index()
print(recommendation_df)
```

```python
movies_to_be_recommend = recommendation_df[recommendation_df["weighted_rating"] > 3.5].sort_values("weighted_rating", ascending=False).head(10)
print(movies_to_be_recommend)
```

```python
## Kodların Fonksiyon Haline Getirilmesi:
import pandas as pd

def create_user_movie_df():
  movie = pd.read_csv("/content/movie.csv")
  rating = pd.read_csv("/content/rating.csv")
  df = movie.merge(rating, how="left", on="movieId")
  comment_counts = pd.DataFrame(df["title"].value_counts())
  rare_movies = comment_counts[comment_counts["count"] <= 1000].index
  common_movies = df[~df["title"].isin(rare_movies)]
  user_movies_df = common_movies.pivot_table(index=["userId"], columns=["title"], values="rating")
  return user_movies_df

user_movies_df = create_user_movie_df()

def user_based_recommender(random_user, user_movies_df, ratio=60, cor_th=0.65, score=3.5):
  random_user_df = user_movies_df[user_movies_df.index == random_user]
  movies_watched = random_user_df.columns[random_user_df.notna().any()].tolist()
  movies_watched_df = user_movies_df[movies_watched]
  user_movies_count = movies_watched_df.T.notnull().sum()
  user_movies_count = user_movies_count.reset_index()
  user_movies_count.columns = ["userId", "movie_count"]
  perc = len(movies_watched) * ratio / 100
  users_same_movie = user_movies_count[user_movies_count["movie_count"] > perc]["userId"]
  final_df = pd.concat([movies_watched_df[movies_watched_df.index.isin(users_same_movie)],
                      random_user_df[movies_watched]])
  corr_df = final_df.T.corr().unstack().sort_values().drop_duplicates()
  corr_df = pd.DataFrame(corr_df, columns=["corr"])
  corr_df.index.names=["user_id_1","user_id_2"]
  corr_df = corr_df.reset_index()
  top_user = corr_df[(corr_df["user_id_1"] == random_user) & (corr_df["corr"] >= cor_th)][["user_id_2", "corr"]].reset_index(drop=True)
  top_user = top_user.sort_values("corr", ascending=False)
  top_user.rename(columns={"user_id_2": "userId"}, inplace=True)
  top_user_rating = top_user.merge(rating[["userId","movieId","rating"]], how="inner")
  top_user_rating = top_user_rating[top_user_rating["userId"] != random_user]
  top_user_rating["weighted_rating"] = top_user_rating["corr"] * top_user_rating["rating"]
  recommendation_df =top_user_rating.groupby("movieId").agg({"weighted_rating": "mean"})
  recommendation_df = recommendation_df.reset_index()
  movies_to_be_recommend = recommendation_df[recommendation_df["weighted_rating"] > score].sort_values("weighted_rating", ascending=False).head(10)
  return  movies_to_be_recommend.merge(movie[["movieId","title"]])
```

```python
random_user = int(pd.Series(user_movies_df.index).sample(1).values[0])
print(random_user)
user_based_recommender(random_user, user_movies_df)
```
