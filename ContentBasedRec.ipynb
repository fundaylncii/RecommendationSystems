```python
import pandas as pd
pd.set_option("display.max_columns", None)
pd.set_option("display.width",500)
pd.set_option("display.expand_frame_repr", False)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df_ = pd.read_csv("/content/movies_metadata.csv", low_memory=False, nrows=25000)
df = df_.copy()
df.head()
df.shape

```

```python

tfidf = TfidfVectorizer(stop_words="english")

df[df["overview"].isnull()]
df["overview"] = df["overview"].fillna(" ")
df.head(20)
```

```python

tfidf_matrix = tfidf.fit_transform(df["overview"])
tfidf_matrix.shape

```

```python
tfidf.get_feature_names_out()
```

```python
tfidf_matrix.toarray()
```

```python
cosinesim = cosine_similarity(tfidf_matrix,tfidf_matrix)
cosinesim.shape
```

```python

################################################################3
## BENZERLİKLERE GÖRE ÜRÜN ÖNERİLERİNİN YAPILMASI

indices = pd.Series(df.index, index=df["title"])
print(indices)

```

```python
indices.index.value_counts()
indices["Hamlet"]
## film isimlerinde çoklamalar var

```

```python
## tekrar verilerinden son olan güncel değerler bırakılarak tekrarlar siliniyor
indices = indices[~ indices.index.duplicated(keep="last")]
indices["Hamlet"]
indices.index.value_counts()
```

```python
pd.set_option("display.float_format", lambda x: "%.5f" % x)
movie_index = indices["Hamlet"]
cosinesim[movie_index]
```

```python
similarity_score = pd.DataFrame(cosinesim[movie_index], columns= ["score"])
print(similarity_score)
```

```python
## Hamlet filmi ile en yüksek benzerliğe sahip filmlerin listesi
movie_indeces = similarity_score.sort_values("score", ascending=False)[1:11].index
print(movie_indeces)
```

```python
df["title"].iloc[movie_indeces]
```

```python
def content_based_recommender(title, cosine_sim, dataframe):
    indices = pd.Series(df.index, index=df["title"])
    indices = indices[~ indices.index.duplicated(keep="last")]
    movie_index = indices[title]
    similarity_score = pd.DataFrame(cosinesim[movie_index], columns=["score"])
    movie_indeces = similarity_score.sort_values("score", ascending=False)[1:11].index
    return dataframe["title"].iloc[movie_indeces]

content_based_recommender("The Matrix", cosinesim, df)
```
